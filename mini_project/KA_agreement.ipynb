```python
# define the setup
files = [
    '/content/drive/MyDrive/sentiment_analysis_data/phase1_annotation/snippets_20_annotatedByHuman.csv',
    '/content/drive/MyDrive/sentiment_analysis_data/phase1_annotation/snippets_20_llmannotated_v4.csv',
]
column_name = 'sentiment'

# define the string-to-integer mapping
mapping_dict = {
    'positive': 1,
    'neutral': 2,
    'negative': 3
}
```

```python
# process the CSVs
rows = []
for file in files:
    # read only the specific column
    df = pd.read_csv(file, usecols=[column_name])

    # map the strings to integers.
    # any string not in the dictionary will become NaN.
    mapped_data = df[column_name].map(mapping_dict)

    # convert the pandas series to a 1D numpy array and add to our list
    rows.append(mapped_data.to_numpy())

# stack the three arrays vertically into a 3 row numpy array
reliability_data = np.vstack(rows)
```

```python
# calculate KA
alpha = krippendorff.alpha(reliability_data, level_of_measurement="nominal")
print(f"Krippendorff's Alpha: {alpha:.3f}")
```
